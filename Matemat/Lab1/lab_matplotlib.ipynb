# üìä –õ–∞–±–æ—Ä–∞—Ç–æ—Ä–Ω–∞ —Ä–æ–±–æ—Ç–∞: –í—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—è –¥–∞–Ω–∏—Ö –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º Matplotlib
–£ —Ü—å–æ–º—É –Ω–æ—É—Ç–±—É—Ü—ñ –º–∏ —Ä–æ–∑–≥–ª—è–Ω–µ–º–æ –ø—Ä–∏–∫–ª–∞–¥–∏ –ø–æ–±—É–¥–æ–≤–∏ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤ –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é –±—ñ–±–ª—ñ–æ—Ç–µ–∫–∏ `matplotlib`. –í—Å—ñ –∫—Ä–æ–∫–∏ —Å—É–ø—Ä–æ–≤–æ–¥–∂—É—é—Ç—å—Å—è –ø–æ—è—Å–Ω–µ–Ω–Ω—è–º–∏ —Ç–∞ –ø—Ä–∏–∫–ª–∞–¥–∞–º–∏.

```python
%pip install matplotlib
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

## üìà –ü–æ–±—É–¥–æ–≤–∞ –±–∞–∑–æ–≤–æ–≥–æ –≥—Ä–∞—Ñ—ñ–∫–∞

```python
plt.plot([1, 2, 4, 8])
plt.title('–ë–∞–∑–æ–≤–∏–π –≥—Ä–∞—Ñ—ñ–∫')
plt.xlabel('–Ü–Ω–¥–µ–∫—Å')
plt.ylabel('–ó–Ω–∞—á–µ–Ω–Ω—è')
plt.show()
```

## üî¢ –ì–µ–Ω–µ—Ä–∞—Ü—ñ—è –≤–∏–ø–∞–¥–∫–æ–≤–∏—Ö –∑–Ω–∞—á–µ–Ω—å —Ç–∞ —Å—Ç–∞—Ç–∏—Å—Ç–∏–∫–∞

```python
s = 100
rng = np.random.default_rng(seed=101)
x = rng.integers(0, 5, size=s, endpoint=True)
y = rng.integers(0, 100, size=s, endpoint=True)
print(f"""
X –º–∞—î {x.size} –µ–ª–µ–º–µ–Ω—Ç—ñ–≤, –∑ –Ω–∏—Ö —É–Ω—ñ–∫–∞–ª—å–Ω—ñ {np.unique(x)}
–ü–æ –•: –ú–∞–∫—Å–∏–º—É–º {x.max()}, –ú—ñ–Ω—ñ–º—É–º {x.min()}, –°–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è: {x.mean()}

Y –º–∞—î {y.size} –µ–ª–µ–º–µ–Ω—Ç—ñ–≤, –∑ –Ω–∏—Ö —É–Ω—ñ–∫–∞–ª—å–Ω–∏—Ö {np.unique(y).size}
–ü–æ Y: –ú–∞–∫—Å–∏–º—É–º {y.max()}, –ú—ñ–Ω—ñ–º—É–º {y.min()}, –°–µ—Ä–µ–¥–Ω—î –∑–Ω–∞—á–µ–Ω–Ω—è: {y.mean()}
""")
```

## üîµ –¢–æ—á–∫–æ–≤–∞ –¥—ñ–∞–≥—Ä–∞–º–∞

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('–í–∏–ø–∞–¥–∫–æ–≤—ñ –≤–µ–ª–∏—á–∏–Ω–∏ —Ö —Ç–∞ —É')
plt.plot(np.arange(x.max()+1), y.mean().repeat(x.max()+1), '--')
plt.show()
```

## üìä –ì—ñ—Å—Ç–æ–≥—Ä–∞–º–∞

```python
plt.hist(y)
plt.title('–ì—ñ—Å—Ç–æ–≥—Ä–∞–º–∞ –∑–Ω–∞—á–µ–Ω—å Y')
plt.xlabel('–ó–Ω–∞—á–µ–Ω–Ω—è')
plt.ylabel('–ß–∞—Å—Ç–æ—Ç–∞')
plt.show()
```

## üìâ –î–µ–∫—ñ–ª—å–∫–∞ –≥—Ä–∞—Ñ—ñ–∫—ñ–≤ –Ω–∞ –æ–¥–Ω–æ–º—É –ø–æ–ª—ñ

```python
x = np.linspace(0, 5, 100)
ya = (5 - x)/4
yb = (3 - x)**2
yc = (2 - x)
yd = np.full_like(x, 1)

plt.plot(x, ya, label='y = (5 - x)/4')
plt.plot(x, yb, label='y = (3 - x)**2')
plt.plot(x, yc, label='y = 2 - x')
plt.plot(x, yd, label='y = 1')
plt.title('–í—ñ–∑—É–∞–ª—ñ–∑–∞—Ü—ñ—è –∫—ñ–ª—å–∫–æ—Ö —Ñ—É–Ω–∫—Ü—ñ–π')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
