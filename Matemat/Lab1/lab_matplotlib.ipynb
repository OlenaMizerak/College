```python
import matplotlib.pyplot as plt
import numpy as np
```

## 📈 Побудова базового графіка

```python
plt.plot([1, 2, 4, 8])
plt.title('Базовий графік')
plt.xlabel('Індекс')
plt.ylabel('Значення')
plt.show()
```

## 🔢 Генерація випадкових значень та статистика

```python
s = 100
rng = np.random.default_rng(seed=101)
x = rng.integers(0, 5, size=s, endpoint=True)
y = rng.integers(0, 100, size=s, endpoint=True)
print(f"""
X має {x.size} елементів, з них унікальні {np.unique(x)}
По Х: Максимум {x.max()}, Мінімум {x.min()}, Середнє значення: {x.mean()}

Y має {y.size} елементів, з них унікальних {np.unique(y).size}
По Y: Максимум {y.max()}, Мінімум {y.min()}, Середнє значення: {y.mean()}
""")
```

## 🔵 Точкова діаграма

```python
plt.scatter(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Випадкові величини х та у')
plt.plot(np.arange(x.max()+1), y.mean().repeat(x.max()+1), '--')
plt.show()
```

## 📊 Гістограма

```python
plt.hist(y)
plt.title('Гістограма значень Y')
plt.xlabel('Значення')
plt.ylabel('Частота')
plt.show()
```

## 📉 Декілька графіків на одному полі

```python
x = np.linspace(0, 5, 100)
ya = (5 - x)/4
yb = (3 - x)**2
yc = (2 - x)
yd = np.full_like(x, 1)

plt.plot(x, ya, label='y = (5 - x)/4')
plt.plot(x, yb, label='y = (3 - x)**2')
plt.plot(x, yc, label='y = 2 - x')
plt.plot(x, yd, label='y = 1')
plt.title('Візуалізація кількох функцій')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
